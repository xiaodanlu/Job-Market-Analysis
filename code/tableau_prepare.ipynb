```python
import pandas as pd
data = pd.read_csv("../dataset/USA_dataset.csv")
data.head()
```

## Location Column

```python
print(data['location'].value_counts())
```

## Subset dataset and clean

```python
subset_dataset = data.iloc[:, [0, 1, 2, 3, -1,-2]]
subset_dataset.head()
# get frequency
subset_dataset['company_freq'] = subset_dataset['company_name'].map(subset_dataset['company_name'].value_counts())
subset_dataset['location_freq'] = subset_dataset['location'].map(subset_dataset['location'].value_counts())
subset_dataset['via_freq'] = subset_dataset['via'].map(subset_dataset['via'].value_counts())
subset_dataset.to_csv('../dataset/us_freq.csv', index=False)
```

```python
import pandas as pd
data = pd.read_csv("../dataset/USA_dataset_nolat.csv")
data.head()
```

## clean

```python
subset_dataset = data.dropna(subset=['location'])
subset_dataset['via'] = subset_dataset['via'].str.replace('via', '')
```

```python
subset_dataset = subset_dataset.iloc[:, [0, 1, 2, 3]]
subset_dataset.head()
subset_dataset['company_freq'] = subset_dataset['company_name'].map(subset_dataset['company_name'].value_counts())
subset_dataset['location_freq'] = subset_dataset['location'].map(subset_dataset['location'].value_counts())
subset_dataset['via_freq'] = subset_dataset['via'].map(subset_dataset['via'].value_counts())
subset_dataset.to_csv('../dataset/us_freq_all.csv', index=False)
```
